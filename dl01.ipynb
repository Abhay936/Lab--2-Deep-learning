```python
a = tf.constant([10,20,30])
print(a.dtype, a.shape,tf.rank(a))
```

```python
z = tf.zeros((2, 3))
print(z)
```

```python
o = tf.ones((3, 2))
print(o)
```

```python
f = tf.fill((2, 2), 9)
print(f)
```

```python
a=tf.constant([10,20,30])
b=tf.constant([50,60,70])
c=tf.add(a,b)
print(c)

d=tf.subtract(a,b)
print(d)

e=tf.multiply(a,b)
print(e)

f=tf.divide(a,b)
print(f)

g=tf.pow(a,2)
print(g)
```
